## 1. åˆå§‹åŒ–è¢å¹•

é€™ä¸€æ­¥è·Ÿä¹‹å‰ä¸€æ¨£ï¼Œæˆ‘å€‘éœ€è¦å…ˆå•Ÿå‹•è¢å¹•çš„é©…å‹•ç¨‹å¼ã€‚

````python
from luma.core.interface.serial import i2c
from luma.core.render import canvas
from luma.oled.device import sh1106
import time

# --- åˆå§‹åŒ– I2C èˆ‡ OLED ---
try:
    serial = i2c(port=1, address=0x3C)
    device = sh1106(serial)
    device.clear()
    print("âœ… OLED åˆå§‹åŒ–æˆåŠŸï¼è¢å¹•ç¾åœ¨æ‡‰è©²æ˜¯å…¨é»‘çš„ã€‚")
except Exception as e:
    print(f"âŒ OLED åˆå§‹åŒ–å¤±æ•—: {e}")
````

## 2. æ–‡å­—æ’ç‰ˆ (Text Positioning)

OLED è¢å¹•å°±åƒä¸€å¼µç•«å¸ƒï¼Œæˆ‘å€‘å¯ä»¥ç”¨ **(x, y) åº§æ¨™** ä¾†æ±ºå®šæ–‡å­—è¦å¯«åœ¨å“ªè£¡ã€‚
* **x**: è·é›¢å·¦é‚Šå¤šé 
* **y**: è·é›¢ä¸Šé¢å¤šé 

æˆ‘å€‘çš„è¢å¹•å¾ˆå°ï¼Œå¤§æ¦‚åªèƒ½æ”¾å…©è¡Œå­—ã€‚
* **ç¬¬ä¸€è¡Œ (æ¨™é¡Œ)**ï¼šå»ºè­°æ”¾åœ¨ `y = 10`
* **ç¬¬äºŒè¡Œ (å…§å®¹)**ï¼šå»ºè­°æ”¾åœ¨ `y = 35`

**ç·´ç¿’ 1ï¼š**
è«‹ä¿®æ”¹ä¸‹æ–¹çš„ç¨‹å¼ç¢¼ï¼Œè®“è¢å¹•é¡¯ç¤ºï¼š
* ç¬¬ä¸€è¡Œï¼š`"Game Start"`
* ç¬¬äºŒè¡Œï¼š`"Score: 0"`

````python
# --- ç·´ç¿’ï¼šé¡¯ç¤ºéŠæˆ²ç‹€æ…‹ ---

with canvas(device) as draw:
    # ç•«ä¸€å€‹é»‘è‰²æ–¹å¡Šæ¸…é™¤èƒŒæ™¯
    draw.rectangle(device.bounding_box, outline="black", fill="black")
    
    # é¡¯ç¤ºç¬¬ä¸€è¡Œæ–‡å­— (æ¨™é¡Œ)
    # æç¤ºï¼šä¿®æ”¹å¼•è™Ÿå…§çš„æ–‡å­—
    draw.text(( , ), "_______", fill="white")
    
    # é¡¯ç¤ºç¬¬äºŒè¡Œæ–‡å­— (åˆ†æ•¸)
    # æç¤ºï¼šä¿®æ”¹å¼•è™Ÿå…§çš„æ–‡å­—
    draw.text(( , ), "________", fill="white")
````

### ğŸ’¡ è§£ç­” (é»æ“Šå±•é–‹)

````python
with canvas(device) as draw:
    draw.rectangle(device.bounding_box, outline="black", fill="black")
    draw.text((10, 10), "Game Start", fill="white")
    draw.text((10, 35), "Score: 0", fill="white")
````

## 3. æ‰“é€ é¡¯ç¤ºå·¥å…· (Helper Function)

ä½ æœ‰æ²’æœ‰ç™¼ç¾ï¼Œæ¯æ¬¡è¦é¡¯ç¤ºæ–‡å­—éƒ½è¦å¯« `with canvas...`, `draw.rectangle...` å¥½å¹¾è¡Œç¨‹å¼ç¢¼ï¼Ÿ
å¦‚æœåœ¨éŠæˆ²ä¸­æˆ‘å€‘è¦æ›´æ–°å¹¾åæ¬¡ç•«é¢ï¼Œç¨‹å¼ç¢¼æœƒè®Šå¾—å¾ˆäº‚ï¼

æˆ‘å€‘å¯ä»¥æŠŠé€™äº›é‡è¤‡çš„å‹•ä½œåŒ…è£æˆä¸€å€‹ **å‡½å¼ (Function)**ã€‚ä»¥å¾Œåªè¦å‘¼å«é€™å€‹å‡½å¼ï¼Œä¸€è¡ŒæŒ‡ä»¤å°±èƒ½æ›´æ–°ç•«é¢ï¼
ä½ å¯ä»¥æŠŠ OLED æƒ³åƒæˆä¸€å€‹æœ‰ **å…©è¡Œæ’æ§½** çš„é›»å­çœ‹æ¿ï¼š

* **`line1` (ç¬¬ä¸€è¡Œ)**ï¼šç”¨ä¾†é¡¯ç¤ºç‹€æ…‹ï¼ˆä¾‹å¦‚ï¼š"Ready...", "You Win!"ï¼‰
* **`line2` (ç¬¬äºŒè¡Œ)**ï¼šç”¨ä¾†é¡¯ç¤ºç´°ç¯€ï¼ˆä¾‹å¦‚ï¼š"Score: 0", "Pi: Rock"ï¼‰

é€™æ¨£ä»¥å¾Œæˆ‘å€‘åªè¦å‘¼å« `show_on_oled(..., "æ¨™é¡Œ", "å…§å®¹")` ä¸€è¡ŒæŒ‡ä»¤å°±æå®šäº†ï¼

**ç·´ç¿’ï¼š**
è«‹å®Œæˆä¸‹æ–¹çš„ `show_on_oled` å‡½å¼ã€‚é€™å€‹å‡½å¼åœ¨æˆ‘å€‘ç¬¬5ç« æœƒç”¨åˆ°ã€‚

````python
# --- ç·´ç¿’ 2ï¼šå®šç¾©é¡¯ç¤ºå‡½å¼ ---

# é€™å€‹å‡½å¼æ¥å—å…©å€‹æ–‡å­—åƒæ•¸ï¼šline1 (ç¬¬ä¸€è¡Œ) å’Œ line2 (ç¬¬äºŒè¡Œ)
def show_on_oled(device, line1, line2):
    with canvas(device) as draw:
        # 1. æ¸…é™¤ç•«é¢
        draw.rectangle(device.bounding_box, outline="black", fill="black")
        
        # 2. é¡¯ç¤ºç¬¬ä¸€è¡Œ (ä½¿ç”¨å‚³é€²ä¾†çš„è®Šæ•¸ line1)
        
        # 3. é¡¯ç¤ºç¬¬äºŒè¡Œ (ä½¿ç”¨å‚³é€²ä¾†çš„è®Šæ•¸ line2)

print("å‡½å¼å®šç¾©å®Œæˆï¼")
````

### ğŸ’¡ è§£ç­” (é»æ“Šå±•é–‹)

````python
def show_on_oled(device, line1, line2):
    with canvas(device) as draw:
        draw.rectangle(device.bounding_box, outline="black", fill="black")
        draw.text((10, 10), line1, fill="white")
        draw.text((10, 35), line2, fill="white")

print("å‡½å¼å®šç¾©å®Œæˆï¼")
````

```python
import cv2
import numpy as np
import time
import os
from tflite_runtime.interpreter import Interpreter 
from picamera2 import Picamera2
from IPython.display import display, Image
import ipywidgets as widgets

# è§£ç­”ï¼šå®Œæ•´çµ•å°è·¯å¾‘
MODEL_PATH = "/home/pi/rock_paper_scissors/teachable_machine_model/model.tflite"
LABEL_PATH = "/home/pi/rock_paper_scissors/teachable_machine_model/labels.txt"

# æª¢æŸ¥æª”æ¡ˆæ˜¯å¦å­˜åœ¨
if not os.path.exists(MODEL_PATH) or not os.path.exists(LABEL_PATH):
    print("âŒ éŒ¯èª¤ï¼šæ‰¾ä¸åˆ°æ¨¡å‹æª”æ¡ˆï¼è«‹æª¢æŸ¥è·¯å¾‘æ˜¯å¦æ­£ç¢ºã€‚")
else:
    print("âœ… æª”æ¡ˆè·¯å¾‘ç¢ºèªç„¡èª¤ã€‚")
```

## ğŸ§  ç·´ç¿’ï¼šæ–‡å­—é­”è¡“å¸« (String Split)

```python
raw_label = "0 Rock"

# ç¬¬ä¸€æ­¥ï¼šåˆ‡é–‹
parts = raw_label.split()
# æ­¤æ™‚ parts è®Šæˆäº† ['0', 'Rock']

# ç¬¬äºŒæ­¥ï¼šæ‹¿ç¬¬ 1 å€‹ (ä¹Ÿå°±æ˜¯ "Rock")
clean_label = parts[1]

print(f"è™•ç†å¾Œçš„æ–‡å­—: {clean_label}")
```

## 4. è¼‰å…¥æ¨¡å‹èˆ‡æ¨™ç±¤
é€™æ®µç¨‹å¼ç¢¼è² è²¬å°‡ `model.tflite` è®€å…¥è¨˜æ†¶é«”ï¼Œä¸¦æº–å‚™å¥½é€²è¡Œé‹ç®—ã€‚æˆ‘å€‘ä¹Ÿæœƒè®€å– `labels.txt`ï¼Œé€™æ¨£ç¨‹å¼æ‰çŸ¥é“ "0" ä»£è¡¨çŸ³é ­ã€"1" ä»£è¡¨å‰ªåˆ€...

```python
def load_labels(path):
    """è®€å–æ¨™ç±¤æª”æ¡ˆï¼Œç§»é™¤æ›è¡Œç¬¦è™Ÿ"""
    with open(path, 'r') as f:
        return [line.strip() for line in f.readlines()]

def load_model(path):
    """è¼‰å…¥ TFLite æ¨¡å‹ä¸¦åˆ†é…å¼µé‡ (Tensors)"""
    interpreter = Interpreter(model_path=path)
    interpreter.allocate_tensors()
    return interpreter

# åŸ·è¡Œè¼‰å…¥
try:
    labels = load_labels(LABEL_PATH)
    interpreter = load_model(MODEL_PATH)
    
    # å–å¾—æ¨¡å‹è¼¸å…¥å’Œè¼¸å‡ºçš„è©³ç´°è³‡è¨Š
    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()
    
    # æŸ¥çœ‹æ¨¡å‹éœ€è¦çš„è¼¸å…¥åœ–ç‰‡å¤§å°
    height = input_details[0]['shape'][1]
    width = input_details[0]['shape'][2]
    
    print(f"âœ… æ¨¡å‹è¼‰å…¥æˆåŠŸï¼")
    print(f"   é¡åˆ¥æ¨™ç±¤: {labels}")
    print(f"   æ¨¡å‹è¼¸å…¥å°ºå¯¸: {width}x{height}")
    
except Exception as e:
    print(f"âŒ æ¨¡å‹è¼‰å…¥å¤±æ•—: {e}")
```

## 5. å®šç¾©å½±åƒè™•ç†èˆ‡æ¨è«–å‡½å¼
é€™æ˜¯ AI çš„æ ¸å¿ƒé‚è¼¯ã€‚ç›¸æ©Ÿæ‹åˆ°çš„ç…§ç‰‡ä¸èƒ½ç›´æ¥çµ¦ AI åƒï¼Œå¿…é ˆç¶“éã€Œé è™•ç†ã€ï¼š
1.  **ç¸®æ”¾ (Resize)**ï¼šè®Šæˆ 224x224 åƒç´ ã€‚
2.  **æ“´å……ç¶­åº¦ (Expand Dims)**ï¼šè®Šæˆ (1, 224, 224, 3) çš„æ ¼å¼ã€‚
3.  **æ¨è«– (Inference)**ï¼šå• AI é€™æ˜¯ä»€éº¼ï¼Ÿ

```python
def process_and_predict(image, interpreter, input_details, output_details):
    """
    1. é è™•ç†å½±åƒ
    2. åŸ·è¡Œæ¨è«–
    3. å›å‚³é æ¸¬çµæœ
    """
    # --- A. é è™•ç† ---
    # å–å¾—æ¨¡å‹éœ€è¦çš„å°ºå¯¸ (224x224)
    input_shape = input_details[0]['shape']
    target_height, target_width = input_shape[1], input_shape[2]
    
    # ç¸®æ”¾å½±åƒ
    resized_image = cv2.resize(image, (target_width, target_height))
    
    # æª¢æŸ¥æ¨¡å‹æ˜¯å¦éœ€è¦æ­£è¦åŒ– (float32) æˆ–ä¿æŒæ•´æ•¸ (uint8)
    # é‡åŒ–æ¨¡å‹ (Quantized) é€šå¸¸ä½¿ç”¨ uint8ï¼Œä¸éœ€è¦é™¤ä»¥ 255
    input_data = np.expand_dims(resized_image, axis=0)
    
    if input_details[0]['dtype'] == np.float32:
        # å¦‚æœæ˜¯æµ®é»æ•¸æ¨¡å‹ï¼Œé€šå¸¸éœ€è¦æ­£è¦åŒ–åˆ° -1 ~ 1 ä¹‹é–“
        input_data = (np.float32(input_data) - 127.5) / 127.5

    # --- B. åŸ·è¡Œæ¨è«– (Inference) ---
    # å°‡è™•ç†å¥½çš„åœ–ç‰‡æ”¾å…¥æ¨¡å‹è¼¸å…¥ç«¯
    interpreter.set_tensor(input_details[0]['index'], input_data)
    
    # é–‹å§‹è¨ˆç®—
    interpreter.invoke()
    
    # --- C. å–å¾—çµæœ ---
    # å¾è¼¸å‡ºç«¯æ‹¿åˆ°é æ¸¬æ•¸æ“š (ä¸€çµ„ä¿¡å¿ƒæŒ‡æ•¸ï¼Œä¾‹å¦‚ [10, 240, 5, 0])
    output_data = interpreter.get_tensor(output_details[0]['index'])[0]
    
    # æ‰¾å‡ºæ•¸å€¼æœ€å¤§çš„é‚£å€‹ç´¢å¼• (ä¾‹å¦‚ index 1 ä»£è¡¨å‰ªåˆ€)
    max_index = np.argmax(output_data)
    
    # è¨ˆç®—ä¿¡å¿ƒç™¾åˆ†æ¯” (å¦‚æœæ˜¯ uint8ï¼Œæ•¸å€¼æ˜¯ 0-255ï¼Œæ‰€ä»¥é™¤ä»¥ 255)
    confidence = output_data[max_index]
    if output_details[0]['dtype'] == np.uint8:
         confidence = confidence / 255.0
            
    return max_index, confidence
```

## 6. ä¸»ç¨‹å¼ï¼šå•Ÿå‹•ç›¸æ©Ÿä¸¦é–‹å§‹è¾¨è­˜

```python
# é¡¯ç¤ºå½±åƒçš„å€åŸŸ (åªä¿ç•™å½±åƒ Widget)
image_widget = widgets.Image(format='jpeg', width=640, height=480)
display(image_widget)

print("ğŸ‘‡ ç¨‹å¼åŸ·è¡Œä¸­...")
print("å¦‚æœè¦çµæŸç¨‹å¼ï¼Œè«‹æŒ‰ä¸Šæ–¹å·¥å…·åˆ—çš„ â¹ (åœæ­¢) æŒ‰éˆ•")

def main_loop():
    print("ğŸ“· ç›¸æ©Ÿå•Ÿå‹•ä¸­...")
    
    try:
        # ä½¿ç”¨ with èªæ³•ç¢ºä¿å®‰å…¨é–‹é—œç›¸æ©Ÿ
        with Picamera2() as picam2:
            # è¨­å®šç›¸æ©Ÿ
            config = picam2.create_preview_configuration(main={"size": (640, 480)})
            picam2.configure(config)
            picam2.start()
            
            # æš–æ©Ÿ
            time.sleep(2)
            print("ğŸš€ AI è¾¨è­˜é–‹å§‹ï¼è«‹å°è‘—é¡é ­å‡ºæ‹³ï¼")

            # æ”¹ç‚ºç„¡é™è¿´åœˆï¼Œç›´åˆ°è¢«æ‰‹å‹•ä¸­æ–·
            while True:
                # 1. æŠ“å–å½±åƒ
                frame = picam2.capture_array()
                
                # 2. è½‰æˆ OpenCV æ ¼å¼ (RGBA -> BGR) ä¸¦ç¿»è½‰
                frame_bgr = cv2.cvtColor(frame, cv2.COLOR_RGBA2BGR)
                frame_flipped = cv2.flip(frame_bgr, 1) # å¦‚æœé¡é ­å€’ç½®è«‹ç”¨ -1ï¼Œå¦å‰‡ç”¨ 1 æˆ– 0
                
                # 3. æŠŠåœ–ç‰‡è½‰æˆ RGB (AI éœ€è¦ RGB)
                frame_rgb = cv2.cvtColor(frame_flipped, cv2.COLOR_BGR2RGB)
                
                # 4. å‘¼å« AI é€²è¡Œé æ¸¬
                class_index, confidence = process_and_predict(frame_rgb, interpreter, input_details, output_details)
                
                # å–å¾—é æ¸¬åç¨±
                prediction_text = labels[class_index].split()[-1] 
                
                # 5. å°‡çµæœç•«åœ¨è¢å¹•ä¸Š
                cv2.rectangle(frame_flipped, (0, 0), (250, 60), (0, 0, 0), -1)
                
                # é¡¯ç¤ºé¡åˆ¥åç¨± (ç¶ è‰²å­—)
                cv2.putText(frame_flipped, f"{prediction_text}", (10, 35), 
                            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
                
                # é¡¯ç¤ºä¿¡å¿ƒæŒ‡æ•¸ (ç™½è‰²å­—)
                cv2.putText(frame_flipped, f"Conf: {confidence:.2f}", (10, 55), 
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

                # 6. æ›´æ–° Jupyter ç•«é¢
                _, buffer = cv2.imencode('.jpeg', frame_flipped)
                image_widget.value = buffer.tobytes()
                
                # ç¨å¾®ä¼‘æ¯ä¸€ä¸‹
                time.sleep(0.05)
                
    except KeyboardInterrupt:
        # ç•¶ä½ æŒ‰ä¸‹ Jupyter çš„åœæ­¢æŒ‰éˆ•æ™‚ï¼ŒæœƒåŸ·è¡Œé€™è£¡
        print("ğŸ›‘ ä½¿ç”¨è€…æ‰‹å‹•åœæ­¢ç¨‹å¼")
        
    except Exception as e:
        print(f"âŒ ç™¼ç”ŸéŒ¯èª¤: {e}")
        
    finally:
        print("ğŸ›‘ ç¨‹å¼å·²çµæŸï¼Œç›¸æ©Ÿè³‡æºå·²é‡‹æ”¾ã€‚")

# åŸ·è¡Œä¸»è¿´åœˆ
main_loop()
```
